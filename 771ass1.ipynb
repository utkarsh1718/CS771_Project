```python
import numpy as np
from sklearn.svm import LinearSVC

def my_map(X):
    X_bipolar = 1.0 - 2.0 * np.asarray(X, dtype=np.float64)
    even = X_bipolar[:, 0::2]
    odd = X_bipolar[:, 1::2]
    n = X_bipolar.shape[0]
    cross = (even[:, :, None] * odd[:, None, :]).reshape(n, -1)
    return np.concatenate([X_bipolar, cross], axis=1)

def my_params(X_map=None, X_raw=None, y=None):
    return {
        "loss": "squared_hinge",
        "C": 5.0,
        "tol": 1e-4,
        "penalty": "l2",
        "dual": False,
        "max_iter": 5000,
    }


train_data = np.loadtxt("public_trn[1].txt")
test_data = np.loadtxt("public_tst[1].txt")

# Extract features and labels
X_train_raw = train_data[:, :-1]
y_train = train_data[:, -1]

X_test_raw = test_data[:, :-1]
y_test = test_data[:, -1]

print(f"   Train shape: {X_train_raw.shape} | Labels: {y_train.shape}")
print(f"   Test shape:  {X_test_raw.shape} | Labels: {y_test.shape}")

print("\n2. Mapping features (32D -> 288D)...")
X_train_mapped = my_map(X_train_raw)
X_test_mapped = my_map(X_test_raw)

print("\n3. Training the LinearSVC model...")
clf = LinearSVC(**my_params())
clf.fit(X_train_mapped, y_train)

print("\n4. Evaluating Accuracy...")
train_acc = clf.score(X_train_mapped, y_train)
test_acc = clf.score(X_test_mapped, y_test)

print(f"   Training Accuracy: {train_acc * 100:.4f}%")
print(f"   Testing Accuracy:  {test_acc * 100:.4f}%")
```
